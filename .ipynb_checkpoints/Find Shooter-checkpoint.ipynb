```python
# Function to draw basketball court
def create_court(ax, color, three_d=False):
    # Short corner 3PT lines
    ax.plot([-220, -220], [0, 140], linewidth=2, color=color)
    ax.plot([220, 220], [0, 140], linewidth=2, color=color)
    
    # 3PT Arc
    three_arc = mpl.patches.Arc((0, 140), 440, 315, theta1=0, theta2=180, facecolor='none', edgecolor=color, lw=2)
    if three_d:
        art3d.pathpatch_2d_to_3d(three_arc, z=0,)
    
    ax.add_artist(three_arc)
    
    
    # Lane and Key
    ax.plot([-80, -80], [0, 190], linewidth=2, color=color)
    ax.plot([80, 80], [0, 190], linewidth=2, color=color)
    ax.plot([-60, -60], [0, 190], linewidth=2, color=color)
    ax.plot([60, 60], [0, 190], linewidth=2, color=color)
    ax.plot([-80, 80], [190, 190], linewidth=2, color=color)
    key_circ =mpl.patches.Circle((0, 190), 60, facecolor='none', edgecolor=color, lw=2)
    if three_d:
        art3d.pathpatch_2d_to_3d(key_circ, z=0,)
    ax.add_artist(key_circ)
    
    # Rim
    rim_circ = mpl.patches.Circle((0, 60), 15, facecolor='none', edgecolor=color, lw=2)
    if three_d:
        art3d.pathpatch_2d_to_3d(rim_circ, z=0,)
    ax.add_artist(rim_circ)

    # Backboard
    ax.plot([-30, 30], [40, 40], linewidth=2, color=color)
    # Remove ticks
    ax.set_xticks([])
    ax.set_yticks([])

    # Set axis limits
    ax.set_xlim(-250, 250)
    ax.set_ylim(0, 470)
    
    
```

```python
def make_position_dataframe(event_data, target_team_id, target_player_id, target_event_id=None):
    
    event_df = pd.DataFrame()
    
    for event in event_data['events']:
        event_id = event['eventId']

        if target_event_id is not None and event_id != str(target_event_id):
            continue
        
        print(event_id)
        
        moment_df = pd.DataFrame()
        
        for moment in event['moments']:
            timestamp = moment[1]
            qtr = moment[0]
            qtr_rem = moment[2]
            shot_rem = moment[3]
            
            for player_pos_data in moment[5]:
                team_id = player_pos_data[0]
                player_id = player_pos_data[1]
                
                
                #if team_id ==target_team_id:
                    #import pdb; pdb.set_trace()
                
                if team_id==target_team_id and player_id==target_player_id:
                    
                    x = player_pos_data[2]
                    y = player_pos_data[3]
                    z = player_pos_data[4]
                    
                    row = {'eventId':event_id, 'timestamp':timestamp,
                          'qtr':qtr, 'qtr_rem':qtr_rem, 'shot_rem':shot_rem,
                          'team_id':team_id, 'player_id':player_id,
                          'x':x, 'y':y, 'z':z}
                    
                    moment_df = pd.concat([moment_df, pd.DataFrame([row])])
                    
            
        event_df = pd.concat([event_df, moment_df])
        
    return event_df
            
            
        
        
```

```python
# Returns 1 if the visiting team is shooting when the x-cord is large, 2 if x-cord is small
def visitor_on_large_x(event):
    # Find the id of the visiting team
    visitor = event['visitor']['teamid']

    player_last = 0
    player_now = 0
    last_x = 0
    now_x = 0
    player_team = 0
    going_forward = 0
    has_possession = 0
    # Loop through the first few moments of the game
    for moment in event['moments']:
        player_last = player_now
        last_x = now_x
        
        # Find the player closest to the ball at this moment
        player_now = find_closest_player(moment)
        
        # Finds the x-cord and team id of the player that's closest
        for i in range(1, 11):
            if player_now == moment[5][i][1]:
                now_x = moment[5][i][2]
                player_team = moment[5][i][0]
        
        # Conditionals to ensure that a player still retains possession
        if player_last == player_now:
            direction = now_x - last_x
            has_possession += 1
        else:
            direction = 0
            has_possession = 0
            going_forward = 0
        
        # Find whether the player with the ball is heading towards a bigger
        # or smaller x coordinate
        if direction > 0:
            going_forward += 1
        elif direction < 0:
            going_forward -= 1
            
        # If a player has had possession for several moments in a row, then
        # they definitely have possession, and we don't need to look for
        # which team has the ball anymore
        if has_possession == 5:
            break
    
    # Returns 1 or 2 depending on whether the player with the ball was on the visiting
    # team and which x direction they were moving in
    if going_forward > 0:
        if player_team == visitor:
            return 1
        else:
            return 2
    else:
        if player_team == visitor:
            return 2
        else:
            return 1
```

```python
#Rim: x = 5.5 y = 25
#rim_x1 = 6.25
#rim_y = 25
#rim_x2 = 87.75
# Function gets the player closest to the ball
# visit_side: The side the visitor team is shooting on. Defaults to 0, which
# has no effect on the function and will calculate closest distance of all players.
# If 1, then visiting team is shooting on the side where the x-coordinate is large.
# If 2, then visiting team is shooting on the side where the x-coordine is small.
def find_closest_player(moment, visit_side=0): 
    # Get coordinates of the ball
    ball_x = moment[5][0][2]
    ball_y = moment[5][0][3]

    # Make a list of all players distance to the ball
    ball_dist = np.zeros(10)
    for i in range(1, 11):
        player_x = moment[5][i][2]
        player_y = moment[5][i][3]
        ball_dist[i - 1] = math.sqrt((player_x - ball_x)**2 + (player_y - ball_y)**2)
    
    # Conditionals that will take out remove all players on the team that
    # doesn't score on the side the ball is currently on
    if visit_side == 1:
        if ball_x > 45:
            ball_dist = ball_dist[5:10]
        else:
            ball_dist = ball_dist[0:5]
    else:
        if ball_x > 45:
            ball_dist = ball_dist[0:5]
        else:
            ball_dist = ball_dist[5:10]
        
    # finds the minimum distance and which person has that distance
    min_value = np.amin(ball_dist)
    min_index = np.where(ball_dist == min_value)
    
    # Get the index of out the double array
    index = min_index[0][0]
    
    player_id = moment[5][index + 1][1]

    return player_id
```
